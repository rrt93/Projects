```python
documents2 = []
documents3 = []
document_stemmed = []

for i in range(0,len(documents)):
    documents2.append(word_tokenize(documents[i]))
    documents3.append(' '.join(documents2[i]).lower())
    document_stemmed.append(''.join(porter.stem(documents3[i])))
```

```python
## Function for computing Jaccardian Similarity
```

```python
def jaccard_similarity(query, document):
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    return len(intersection)/len(union)
```

```python
docnames = ['MIT','Stanford', 'Tesla', 'UIC', 'UIS', 'UIUC']
```

```python
## Computing Jaccardian similarity for all 6 documents
```

```python
js_sim = []
for i in range(0,6):
    for j in range(0,6):
        js = jaccard_similarity(document_stemmed[i],document_stemmed[j])
        js_sim.append(js)
```

```python
js_sim_frame = pd.DataFrame(np.array(js_sim).reshape(6,6), columns = ['MIT','Stanford', 'Tesla', 'UIC', 'UIS', 'UIUC'],
                           index =  ['MIT','Stanford', 'Tesla', 'UIC', 'UIS', 'UIUC'])
```

```python
## Function for highlighting tfidf scores greater than 0.005
```
