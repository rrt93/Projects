```python
EM_data['Channel'] = 'EM'
DM_data['Channel'] = 'DM'
SM_data['Channel'] = 'SM'
```

```python
EM_data2 = EM_data[EM_data['days_to_resp']<=30]
SM_data2 = SM_data[SM_data['days_to_resp']<=30]
DM_data2 = DM_data[DM_data['days_to_resp']<=60]
Digital_data4 = Digital_data3[Digital_data3['days_to_resp']<=30]
ThirdParty_data3 = ThirdParty_data2[ThirdParty_data2['days_to_resp']<=30]
```

```python
ThirdParty_data3.head()
```

```python
comb = pd.concat([EM_data2,SM_data2, DM_data2, Digital_data4, ThirdParty_data3], axis=0)
print(comb.shape)
```

```python
comb2 = comb.sort_values(['CURR_AG_IND_ID', 'first_purch_dt2','drop_date2'],
              ascending = [True, True, True])
```

```python
comb3 = comb2[['CURR_AG_IND_ID','first_purch_dt2','days_to_resp','drop_date2','Channel']]
comb3.head()
```

```python
count1 = comb3.groupby(['CURR_AG_IND_ID','first_purch_dt2'])['Channel'].nunique().reset_index().rename(columns={'Channel':'Unique_channels'})
count1.head()
```

```python
comb4 = pd.merge(comb3, count1, how = 'left', on = ['CURR_AG_IND_ID','first_purch_dt2'])
comb4.head()
```

```python
comb4.groupby('Channel')['CURR_AG_IND_ID'].nunique()
```

```python
print(comb4.shape)
comb5 = comb4[comb4['Unique_channels'] > 1]
print(comb5.shape)
comb5.head()
```
